```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./../data/hotel_2021_11_29_vol6.xlsx')

df
```

```python
df.columns
```

```python
# 5, 5성 > 5성급

df = df.replace({
    '현행 호텔등급': {'5성': '5성급', '5': '5성급', '4성': '4성급', '4': '4성급', '3': '3성급', '3성': '3성급', '2성': '2성급', '2': '2성급',
             '1성': '1성급', '1': '1성급', '1등급': '1성급', '2등급': '2성급'}
})
```

```python
df.shape
```

```python
# 영업상태(19.12.31기준) - , 현행 호텔등급 - 5, 4, 3, 2, 1성급

df = df[(df['현행 호텔등급'].isin(['5성급', '4성급', '3성급', '2성급', '1성급'])) & (df['영업상태(19.12.31기준)'].isin(['영업중']))]
```

```python
df.shape   # (723, 27) > 영업상태가 영업중이고, 현행 호텔 등급이 1~5성급 안에 드는 것
```

```python
group_df = df.groupby(['시도', '현행 호텔등급']).head(5).sort_values(by=['시도', '현행 호텔등급'], ascending=[True, False])
```

```python
group_df
```

```python
group_df.to_csv('./../data/시도별_호텔등급_확인.csv', encoding='utf-8-sig')
```
