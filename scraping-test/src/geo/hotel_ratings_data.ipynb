```python
area_dic = {
    '전북': '전라북도',
    '전남': '전라남도',
    '대전': '대전광역시',
    '대구': '대구광역시',
    '서울디특별시': '서울특별시',
    '서울특별시': '서울특별시',
    '강원': '강원도',
    '울산': '울산광역시',
    '대전': '대전광역시',
    '경상북도 ': '경상북도',
    '경상북도': '경상북도',
    '광주': '광주광역시'
}

def changing_area(x):
    if x in area_dic.keys():
        x = area_dic[x]
        return x
    return x
```

```python
df_2018['지역1(시도)'] = df_2018['지역1(시도)'].apply(changing_area)
```
