```python
import pandas as pd
import numpy as np
```

```python
hotel = pd.read_excel('./../data/hotel_list_origin.xlsx', header=1)
```

```python
hotel
```

```python
hotel.loc[:9]
```

```python
hotel.columns = ['idx', 'addr1', 'addr2', 'type', 'name', 'biz_name', 'status', 'rate', 'rate_date', 'floor', 'area1', 'area2', 'rooms', 'address', 'amenities']
```

```python
hotel
```

```python
# 5, 5성 > 5성급

df = hotel.replace({
    'rate': {'5성': '5성급', '5': '5성급', '4성': '4성급', '4': '4성급', '3': '3성급', '3성': '3성급', '2성': '2성급', '2': '2성급',
             '1성': '1성급', '1': '1성급', '1등급': '1성급', '2등급': '2성급'}
})
```

```python
# status(일일상태) - 휴업, rate(등급) - 5, 4, 3, 2, 1성급
df =  df[(df['rate'].isin(['5성급', '4성급', '3성급', '2성급', '1성급'])) & (df['status'].isin(['영업중']))]
```

```python
# 723 rows, 15 cols
df.shape
```

```python
group_df = df.groupby(['addr1', 'rate']).head(5).sort_values(by=['addr1', 'rate'], ascending=[True, False])
```

```python
# rows : 286

group_df[['addr1', 'addr2', 'rate', 'type', 'rooms']].to_csv('시도별_호텔등급확인_20211124.csv', encoding='utf-8-sig')
```
