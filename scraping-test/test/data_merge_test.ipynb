```python
df = org_df.drop([0, 1]).reset_index(drop=True)
```

```python
df.columns = ['idx', 'addr1', 'addr2', 'type', 'name', 'biz_name', 'status', 'rate', 'rate_date', 'floor', 'area1', 'area2', 'rooms', 'address', 'amenities']
```

```python
to_df = pd.read_excel('./../data/hotel_list_20211122.xlsx')
to_df
```

```python
merge_df = pd.merge(left=df, right=to_df, how='left', left_on=['idx', 'name'], right_on=['ID', '호텔명'])
```

```python
merge_df.drop(['ID', '호텔명', '업종'], axis=1)
```
